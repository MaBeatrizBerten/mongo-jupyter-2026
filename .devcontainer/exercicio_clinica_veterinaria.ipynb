```python
consulta_valor = list(
    colecao_atendimentos.find(
        {"valor": {"$gte": 90, "$lte": 200}},
        {"_id": 0}
    ))

print("Atendimentos com valor entre 90 e 200:")
for item in consulta_valor:
    print(item)
```

```python
cidades = ["Assis", "Ourinhos", "Pompeia", "Marília"]
consulta_cidades = list(
    colecao_atendimentos.find(
        {"cidade": {"$in": cidades}},
        {"_id": 0}
    ))

print("\nAtendimentos das cidades específicas:")
for item in consulta_cidades:
    print(item)
```

```python
df = pd.DataFrame(consulta_cidades)

print("\nDataFrame:")
print(df.head())
```

```python
print("\nValores nulos por coluna:")
print(df.isnull().sum())
```

```python
faturamento_total = float(df["valor"].sum())

faturamento_por_cidade = (
    df.groupby("cidade")["valor"]
    .sum()
    .astype(float)
    .to_dict())

faturamento_por_servico = (
    df.groupby("servico")["valor"]
    .sum()
    .astype(float)
    .to_dict())

ticket_medio_geral = float(df["valor"].mean())

ticket_medio_por_cidade = (
    df.groupby("cidade")["valor"]
    .mean()
    .astype(float)
    .to_dict())

servico_mais_lucrativo = (
    df.groupby("servico")["valor"]
    .sum()
    .idxmax())

cidade_maior_faturamento = (
    df.groupby("cidade")["valor"]
    .sum()
    .idxmax())

top_3_atendimentos = (
    df.sort_values(by="valor", ascending=False)
    .head(3)
    .to_dict(orient="records"))
```

```python
relatorio = {
    "data_geracao": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "faturamento_total": faturamento_total,
    "faturamento_por_cidade": faturamento_por_cidade,
    "faturamento_por_servico": faturamento_por_servico,
    "ticket_medio_geral": ticket_medio_geral,
    "ticket_medio_por_cidade": ticket_medio_por_cidade,
    "servico_mais_lucrativo": servico_mais_lucrativo,
    "cidade_maior_faturamento": cidade_maior_faturamento,
    "top_3_atendimentos": top_3_atendimentos
}

print("\nRelatório gerado:")
print(relatorio)
```

```python
colecao_relatorios.insert_one(relatorio)

print("\nRelatório salvo na coleção 'relatorios' com sucesso!")
```
